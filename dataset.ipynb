```python
import numpy as np

def get_random_ua():

    random_ua = ''

    ua_file = 'ua_file.txt'

    try:

        with open(ua_file) as f:

            lines = f.readlines()

        if len(lines) > 0:

            prng = np.random.RandomState()

            index = prng.permutation(len(lines) - 1)

            idx = np.asarray(index, dtype=np.integer)[0]

            random_proxy = lines[int(idx)]

    except Exception as ex:

        print('Exception in random_ua')

        print(str(ex))

    finally:

        return random_ua
```

```python
#size = []
#lista_t = []
#label_list = []
#attr_list = []
def parser_attr(lista_url):
    i = 0
    for page in range(len(lista_url)):
        url = lista_url[i]
        headers = {'user-agent': 'Mozilla/5.0'}
        response = requests.get(url, headers=headers)
        html_soup = BeautifulSoup(response.text, 'html.parser')
        #link = html_soup.find_all('li', class_ = 'd-flex align-items-center col-12 col-sm-6 p-0 m-0 pb-sm pr-xxsm')#
        #label = html_soup.find_all('label', class_ = 'css-1v9f2og ecl3kjh0')#
        attr = html_soup.find_all('div', class_ = 'css-19hiur5 css-dwl48b css-1cnqmgc')#
        #for x in link:
        #    lista_t.append(x)
            #size.append(x.text)
        #for z in label:
        #    label_list.append(z.text)
        j = 0
        for x in attr:
            #attr_list.append(x.text)
            if j==0:
                sede.append(x.text)
            elif j ==1:
                tamanho.append(x.text) # necessario fazer limpeza dos dados
            elif j ==2:
                fundação.append(x.text)

            elif j ==3:
                tipo.append(x.text)
            elif j ==4:
                setor.append(x.text)
            else:
                receita.append(x.text)

            j = j+1
        time.sleep(1)#20 horas+- (60)      
        if page%100 == 0:
            time.sleep(10) 
        i = i +1
```
